# Projectile Game

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import random

#random wall height & location
vx = random.randint(-10, 10)
vy = random.randint(-10, 10)

def parabola(a, b, c, zoom): 
    
    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom
    points = 10*xmax-xmin
    x = np.linspace(xmin, xmax, points)

    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'k') # black x axis
    plt.plot([0,0],[ymin,ymax], 'k') # black y axis

    # plot point/wall
    plt.plot([vx],[vy],'ro') 

    #plot parabola 
    y = a*x**2 + b*x + c  
    plt.plot(x, y)

    plt.show()

#sliders
interactive_plot = interactive(parabola, a=(-100, 100),b=(-100, 100), c=(-100, 100), zoom=(1,100))
interactive_plot
```
